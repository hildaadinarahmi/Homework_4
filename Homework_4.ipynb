```python
import pandas as pd
```

```python
# file_path = 'Homework-DataFrame-Dataset.csv'  #update path if needed

# data = pd.read_csv('Homework-DataFrame-Dataset.csv', sep=';', on_bad_lines='skip')
# data.head()

tempat_file = 'Homework-DataFrame-Dataset.xlsx'
paid_transaction = pd.read_excel(tempat_file, sheet_name='Paid-Transaction')
seller = pd.read_excel(tempat_file, sheet_name='Seller')

seller.head()
```

```python
#no1
paid_transaction['Full Name'] = (paid_transaction['First Name'] + '  ' + paid_transaction['Last Name']).str.title()

cols = paid_transaction.columns.tolist()
last_name_index = cols.index("Last Name")
cols.insert(last_name_index + 1, cols.pop(cols.index("Full Name")))
paid_transaction = paid_transaction[cols]

paid_transaction.head()
```

```python
#no2
paid_transaction['GMV'] = paid_transaction['Transaction Amount'] - paid_transaction ['Delivery Fee']
paid_transaction
```

```python
 #mengkonversikan paid date dari format integer ke format date
# Convert 'Paid Date' from integer format to datetime
paid_transaction['Paid Date'] = pd.to_datetime(paid_transaction['Paid Date'], format = '%Y%m%d', errors = 'coerce')
paid_transaction


```

```python
meta_category_map = {
    'Home/Food' : 'Group 1', 'Sports/Hobi/Otomotif': 'Group 1', 'Fashion': 'Group 1', 'Beauty/Health': 'Group 1',
    'Gadget/Komputer': 'Group 2', 'Elektronik': 'Group 2', 'Service/Mokado': 'Group 2', 'Babies/Kids': 'Group 2'
}
paid_transaction ['Group'] = paid_transaction ['Meta Category'].map(meta_category_map)

pivot_df = paid_transaction.pivot_table(values ='GMV', index=paid_transaction['Paid Date'].dt.to_period('M'), columns ='Group', aggfunc='sum')
paid_transaction
```

```python
august_data = paid_transaction[paid_transaction['Paid Date'].dt.month == 8]
august_data = august_data[august_data['Paid Date'].dt.year == 2017]


august_gmv = august_data.groupby('Order Number') ['GMV'].sum().idxmax()
seller_with_highest_gmv = seller[seller['Order Number'] == august_gmv]
seller_with_highest_gmv
```

```python
september_data = paid_transaction[(paid_transaction['Paid Date'].dt.month == 9) &
                                  (paid_transaction['Paid Date'].dt.year ==2017) &
                                  (paid_transaction['Meta Category']== 'Fashion')]

most_transactions_seller = september_data['Order Number'].value_counts().idxmax()
seller_with_most_transactions = seller[seller['Order Number']== most_transactions_seller]
seller_with_most_transactions
```
